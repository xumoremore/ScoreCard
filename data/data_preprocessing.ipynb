```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

# 数据探索

```python
df = pd.read_csv('cs-training.csv')
pd.set_option("display.max_columns", None)
print(df.columns)
df
```

```python
# 去掉第一列
df.drop('Unnamed: 0', axis='columns', inplace=True)
df.head()
```

```python
# 对违约率做直方图统计，并且计算违约率
# 1:违约，0：没违约
SD = df['SeriousDlqin2yrs'].value_counts()
plt.bar(SD.index, SD.values)
# sns.bar(data=SD)
print(SD)
print("违约率：|{}|".format(SD[1] / len(df['SeriousDlqin2yrs'])))
```

# 数据预处理

```python
df.fillna(df.median())
```

```python
# 对age字段做分箱[ -math.inf, 25, 40, 50, 60, 70, math.inf ], pd.cut(字段，bins)
# 左开右闭
df_bin = pd.DataFrame()
df_bin['SeriousDlqin2yrs'] = df['SeriousDlqin2yrs']
age_bins = [-math.inf, 25, 40, 50, 60, 70, math.inf]
df_bin['bin_age'] = pd.cut(df['age'], bins=age_bins)

# 对'NumberOfDependents'（家属人数）字段分箱[ -math.inf, 2, 4, 6, 8, 10, math.inf]
NumberOfDependents_bins = [-math.inf, 2, 4, 6, 8, 10, math.inf]
df_bin['bin_NumberOfDependents'] = pd.cut(df['NumberOfDependents'], bins=NumberOfDependents_bins)

# 'NumberOfTime30-59DaysPastDueNotWorse', 'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfTimes90DaysLate',
# 分箱[ -math.inf, 1,2,3,4,5,6,7,8,9, math.inf ]
NotWorse_bins = [-math.inf, 1,2,3,4,5,6,7,8,9, math.inf]
df_bin['bin_NumberOfTime30-59DaysPastDueNotWorse'] = pd.cut(df['NumberOfTime30-59DaysPastDueNotWorse'], bins=NotWorse_bins)
df_bin['bin_NumberOfTime60-89DaysPastDueNotWorse'] = pd.cut(df['NumberOfTime60-89DaysPastDueNotWorse'], bins=NotWorse_bins)
df_bin['bin_NumberOfTimes90DaysLate'] = pd.cut(df['NumberOfTimes90DaysLate'], bins=NotWorse_bins)

# 对等频的方法将剩下的五个字段分成5段 duplicates=drop（可能是一个数据放两个箱子里就drop掉一个）
# 'RevolvingUtilizationOfUnsecuredLines', 'DebtRatio', 'MonthlyIncome', 'NumberOfOpenCreditLinesAndLoans', 'NumberRealEstateLoansOrLines',
qcut_vals = ['RevolvingUtilizationOfUnsecuredLines', 'DebtRatio', 'MonthlyIncome', 'NumberOfOpenCreditLinesAndLoans', 'NumberRealEstateLoansOrLines',]
for qcut_val in qcut_vals:
    df_bin['bin_' + qcut_val] = pd.qcut(df[qcut_val], q=5, duplicates="drop")
    
print("df_bin.shape:|{}| \ndf_bin.columns：|{}|".format(df_bin.shape, df_bin.columns))
df_bin.head()
```

# 数据保存

```python
df_bin.to_csv('preprocessed_data.csv', index=False)
```
